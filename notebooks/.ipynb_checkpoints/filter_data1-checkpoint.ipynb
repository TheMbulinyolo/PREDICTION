```python
import pandas as pd
```

```python
path = '../data/valeurs_foncieres/ValeursFoncieres-2025-S1.txt'
```

```python
df = pd.read_csv(
    path,
    sep="|",
    encoding="latin-1",
    low_memory=False)
```

```python
df.head()
```

```python
cols_utiles = [
    "Valeur fonciere",
    "Surface reelle bati",
    "Nombre pieces principales",
    "Code departement",
    "Type local"]
```

```python
df = df[cols_utiles]
```

```python
df.head()
```

```python
for col in ["Valeur fonciere", "Surface reelle bati", "Nombre pieces principales"]:
    df[col] = (df[col].astype(str).str.replace(",",".", regex=False).astype(float))
```

```python
df = df.dropna()
```

```python
df.head()
```

```python
df = df[df["Code departement"] == "75"]
```

```python
df = df[df["Type local"] == "Appartement" ]
```

```python
df
```

```python
df = df[df["Surface reelle bati"] > 9 ]
```

```python
df["prix_m2"] = df["Valeur fonciere"] / df["Surface reelle bati"]
```

```python
df = df[(df["prix_m2"] > 2000) & (df["prix_m2"] < 30000 )]
```

```python
df[["prix_m2"]].describe()
```

```python
outpath = "../data/processed/paris_appart_clean.csv"
```

```python
df.to_csv(
    outpath,
    index=False,
    encoding="utf-8")
```

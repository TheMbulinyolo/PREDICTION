```python
import numpy as np
```

```python
from catboost import CatBoostRegressor
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
```

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed/paris_appart_clean_01.csv")
```

```python
features = [
    "Surface reelle bati",
    "Nombre pieces principales",
    "arrondissement"
]
```

```python
x = df[features]
```

```python
y = np.log(df["prix_m2"])
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = CatBoostRegressor(
    iterations=600,
    depth=8,
    learning_rate=0.05,
    loss_function="RMSE",
    verbose=100
)
```

```python
model.fit(x_train, y_train)
```

```python
preds_log = model.predict(x_test)
```

```python
print("R² (log) :", r2_score(y_test, preds_log))
```

```python
y_test_real = np.exp(y_test)
```

```python
preds_real = np.exp(preds_log)
```

```python
print("MAE €/m²:",mean_absolute_error(y_test_real, preds_real))
```

```python
import joblib
```

```python
joblib.dump(
    {
        "model":model,
        "features":features,
    },
    "../models/prix_m2_model.pkl")
```
